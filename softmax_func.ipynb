```python
import math
```

```python
logits = [1.9,-2.1,-.3, .4, -.8]
```

```python
def softmax(v):
    p = []
    for i in range( len(v)):
        p.append( math.exp(v[i]))
    total = sum(p)
    for i in range(len(v)):
        p[i] = p[i] / total
    return p
```

```python
probs = []
for i in range(len(logits)):
    probs.append(softmax(logits[i]))
    print(probs[i], sum(probs[i]))
```
